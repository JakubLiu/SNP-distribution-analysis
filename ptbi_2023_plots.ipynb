# Within category SNP distribtion

```python
def SNP_dist(path):
    file = open(path, 'r')
    data = file.read()
    data_list = data.split("\n")
    data_list.pop(-1)
    file.close()
    data_set = set(data_list)
    data_uniq = sorted(list(data_set))
    data_uniq_int = []
    counts = []

    for i in data_uniq:
        data_uniq_int.append(int(i))

    for j in data_uniq_int:
        count = data_list.count(str(j))
        counts.append(count)
        
    outlist = [None, None]
    outlist[0] = data_uniq_int
    outlist[1] = counts
    return outlist
```
